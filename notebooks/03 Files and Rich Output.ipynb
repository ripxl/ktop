# Files and Rich Output
A `ktop.Notebook` can load and save `ipynb` files.

```python
import ktop, time
```

```python
nb = ktop.Notebook(file_name="experiments/matplotlib").load()
nb.run_one()
```

The outputs a kernel create are crudely accessible

```python
display(nb.kernels[0].nbformat.cells[-1].outputs[0].data, raw=True)
```

```python
for i in range(10):
    k = nb.run_one(cells=nb.nbformat.cells, reuse=False)
    k.nbformat.cells[2].source = f"y = [math.sin(x * {i}) for x in x]"
```

Wait a sec...

```python
[display(k.nbformat.cells[-1].outputs[0].data, raw=True) for k in nb.kernels];
```
